```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(source,'lxml')
```

```python
#Open a writable csv file, which we will use to store the postcode table
table_csv = open('pc_wiki.csv','w')
csv_writer = csv.writer(table_csv)
csv_writer.writerow(['PostalCode','Borough','Neighborhood'])
```

```python
# Parser postcode table tag from the webpage
table = soup.find('table',class_='wikitable sortable')
#print(table)
```

```python
# Retrive text in the first row of the table
row = table.find('tr')
row_split = row.text.split('\n')
Postcode = row_split[1]
Borough = row_split[2]
Neighbourhood = row_split[3]
print(Postcode,Borough,Neighbourhood)
```

```python
# Retirve text of all rows in the postcode table.
# Write every row of the table to 'pc_wiki.csv' file opened earlier.
# Skip the first row since they are headers.

for row in table.find_all('tr')[1:]:
    row_split = row.text.split('\n')
    Postcode = row_split[1]
    Borough = row_split[2]
    Neighbourhood = row_split[3]
    #print(Postcode,Borough,Neighbourhood)
    csv_writer.writerow([Postcode,Borough,Neighbourhood])
```

```python
# After writing the table, close up the 'pc_wiki.csv' file.
table_csv.close()
```

# Pre-processing the neighborhood table

```python
import random        # library for random number generation
import numpy as np   # library for vectorized computation
import pandas as pd  # library to process data as dataframes
import scipy
```

```python
# Read the 'pc_wiki.csv' file into pandas dataframe.
post_df = pd.read_csv('pc_wiki.csv')
post_df.tail()
```

**Ignore cells with a borough that is 'Not assigned'.**

```python
df = post_df[post_df['Borough'] != 'Not assigned'].reset_index(drop=True)
df.tail()
```

```python
Postcode_list = df['PostalCode'].unique()
print('There are {} distinct post codes'.format(len(Postcode_list)))
```

```python
# Join neighbors with the same post code by comma
df = df.groupby(['PostalCode', 'Borough'])['Neighborhood'].apply(', '.join).reset_index()
df.shape
```

```python
# Neighborhood will be the same as the borough if 'Not assigned'
df['Neighborhood'].replace('Not assigned', "Queen's Park", inplace=True)
df
```

## Utilize the Foursquare API

```python
CLIENT_ID = 'A valid id' # your Foursquare ID
CLIENT_SECRET = 'A valid secrete' # your Foursquare Secret
VERSION = '20190425' # Foursquare API version

#print('Your credentails:')
#print('CLIENT_ID: ' + CLIENT_ID)
#print('CLIENT_SECRET:' + CLIENT_SECRET)
```

```python
# Get the neighborhood's latitude and longitude values.
nbhood_latitude = Toronto_data.loc[0,'Latitude']  # neighborhood latitude value
nbhood_longitude= Toronto_data.loc[0,'Longitude'] # neighborhood longitude value
nbhood_name = Toronto_data.loc[0,'Neighborhood']  # neighborhood name
print('Latitude and longitude values of {} are {}, {}.'.format(nbhood_name,nbhood_latitude,nbhood_longitude))
```

**Now, let's get the top 100 venues that are in 'The Beaches' within a radius of 500 meters.**

```python
radius = 500
LIMIT = 100
url = 'https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}'\
      .format(CLIENT_ID,CLIENT_SECRET, nbhood_latitude, nbhood_longitude, VERSION, radius, LIMIT)
                                                                                                                           
#url 
```

```python
# Send the GET request and examine the resutls
results = requests.get(url).json()
```

```python
# Convert venues to pandas dataframe
venues = results['response']['groups'][0]['items']
nearby_venues = json_normalize(venues) # flatten JSON
```

```python
# Filter out main features of venues
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues = nearby_venues[filtered_columns]
range(nearby_venues.shape[0])
```

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)
```

```python
# clean title of the columns
nearby_venues.columns = [col.split('.')[-1] for col in nearby_venues.columns]
print('{} venues were returned by Foursquare.'.format(nearby_venues.shape[0]))
```

**Repeat the above for each Neighborhood in Toronto**

```python
def getNearbyVunues(nbhood_latitude,nbhood_longitude,nbhood_name,radius,LIMIT):
    
    venues_list = pd.DataFrame()
    
    for name, lat, lng in zip(nbhood_name, nbhood_latitude,nbhood_longitude):
        url = 'https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}'\
               .format(CLIENT_ID,CLIENT_SECRET, lat, lng, VERSION, radius, LIMIT)
        
        results = requests.get(url).json()
        venues = results['response']['groups'][0]['items']
        nearby_venues = json_normalize(venues)
        nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)
        print(name)
        
        # return only relevant information for each nearby venue
        for ind in  range(nearby_venues.shape[0]):
            
            venues_list = venues_list.append({'Neighborhood': name,
                                              'Neighborhood Latitude': lat, 
                                              'Neighborhood Longitude': lng,
                                              'Venue': nearby_venues['venue.name'][ind],
                                              'Venue Latitude':  nearby_venues['venue.location.lat'][ind],
                                              'Venue Longitude': nearby_venues['venue.location.lng'][ind],
                                              'Venue Category':nearby_venues['venue.categories'][ind]}, ignore_index=True)
            
    return(venues_list)
        
    
    
```

```python
# Get venues of Toronto Neighborhood
toronto_venues = getNearbyVunues(Toronto_data['Latitude'],
                                 Toronto_data['Longitude'],
                                 Toronto_data['Neighborhood'],
                                 radius = 500,
                                 LIMIT = 100 )
toronto_venues.shape
```

```python
toronto_venues.head()
```

```python
print('There are {} uniques categories.'.format(len(toronto_venues['Venue Category'].unique())))
```

**Explore venue category in each neighborhood**

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")
# add neighborhood column back to dataframe
toronto_onehot.insert(loc=0, column='AllNeighborhood', value=toronto_venues['Neighborhood'])

toronto_onehot.head()
```

```python
toronto_onehot.shape
```

**Group rows by neighborhood and by taking the mean of the frequency of occurrence of each category**

```python
toronto_grouped = toronto_onehot.groupby('AllNeighborhood').mean().reset_index()
toronto_grouped.head()
```

## Let us see the top 5 venues in each neighborhood

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 5

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['AllNeighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['AllNeighborhood'] = toronto_grouped['AllNeighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

# Cluster Toronto Neighborhoods
**Run k-means to cluster the neighborhood into 5 clusters.**

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('AllNeighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
toronto_merged = Toronto_data.join(neighborhoods_venues_sorted.set_index('AllNeighborhood'), on='Neighborhood')
```

```python
toronto_merged.head()
```

## Finally, let's visualize the resulting clusters

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
